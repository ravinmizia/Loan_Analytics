## Null Values Handling

```python
df1.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df1.isnull(), cbar=False, cmap='viridis')
plt.title('Null Values Heatmap')
plt.show()
```

We see that visually, there are lots of missing values.

We have several options in handling missing value in data.

First, we can just drop or erase any rows containing missing value.

Second, we can do some imputations.

In this case if we drop rows containing missing values, it might distort the dataset and we might as well losing some valuable informations.

So we go with the other alternatives that is to impute values into/replacing the missing values.

```python
# examine the missing values columns
df1.columns[df1.isnull().any()]
```

```python
# calculate percentage of missing values
percent_missing = round((df1.isnull().sum() * 100 / len(df1)),2)
missing_value_df = pd.DataFrame({'column_name': df1.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

### Impute mode for categorical columns

```python
# Create subplots
fig, axes = plt.subplots(1, 6, figsize=(15, 3))

# Define the list of column names
categorical_columns = ['Gender', 'Married', 'Dependents', 'Self_Employed', 'Loan_Amount_Term','Credit_History']

# Iterate over the categorical_columns and plot the pie charts
for i, column in enumerate(categorical_columns):
    # Count the frequency of each category in the column
    counts = df1[column].value_counts()
    labels = counts.index.tolist()
    values = counts.values.tolist()

    # Create the pie chart
    axes[i].pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    axes[i].set_title('{}'.format(column))

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

The proportion of unique value in each categorical column is imbalanced, and also one value tend to dominate e.g. Male 81% and female 18%.

With this reason it seems to be accurate to impute missing value into the table wiith its respective mode value or value that comes out the most from a column.

```python
# Iterate over the columns and impute missing values with mode
for column in categorical_columns:
    mode_value = df1[column].mode()[0]  # Calculate the mode value
    df1[column].fillna(mode_value, inplace=True)  # Fill missing values with mode

# Display the updated DataFrame
df1.isnull().sum()
```

we see that the missing values in categorical columns are handled, now we have still null values in numerical volumn to be handled

### Impute mean for numerical column

```python
# Create subplots
fig= plt.figure(figsize=(5, 5))

# Plot the distplots
sns.histplot(df1['LoanAmount'])
plt.axvline(df1['LoanAmount'].mean(), c='r')
plt.text(160, 80,round(df1['LoanAmount'].mean(),2), fontsize=12,c='r')

# Set the titles and labels
plt.title('Loan Amount')
plt.xlabel('Loan Amount')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()

#print
```

The mean for Loan Amount column is 146.41 so this will be imputed to the missing value.

Impuing with mean is one of the most common way to handle missing value from numerical column, because it will less likely to distort the data distribution. 

Also the data is not skewed to imputing with mean mean is good.

```python
#list numerical columns
numerical_columns = ['LoanAmount']

# Iterate over the columns and impute missing values with mode
for column in numerical_columns:
    mode_value = df1[column].mean()  # Calculate the mean value
    df1[column].fillna(mode_value, inplace=True)  # Fill missing values with mode

# Display the updated DataFrame
df1.isnull().sum()
```

Now all the missing values are already imputed with mean
